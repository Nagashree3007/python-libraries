```python
import matplotlib.pyplot as plt
import pandas as pd
import loggingfile as mlf

logger = mlf.logger_init("lineDifferentStyles_P7.ipynb")

def draw_line(data):

    """
    Description : Function to plot two or more lines with different styles
    Parameters : 
        data : Function that converts the data into a tabular format and then uses this
                data to plot the x-axis and y-axis.
    Returns :
        None :
    """

    print("inside draw_line method")
    
    df = pd.DataFrame(data)
    print(f"\n{df}")

    plt.plot(df['Name'],df['Age'], color='blue', label ='line1-width-3', linestyle='-.',marker='o')
    plt.plot(df['Name'],df['Days'], color='green', label='line1-width-4', linestyle='-', marker='1')
    print("plotting the results")

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    print("labellig the x-axis and y-axis")

    plt.legend()
    plt.grid()
    plt.show()


def main():

    print("insdie main method")

    data = {
            'Name':['suresh','mahesh','tarun','sanju','sitaraman'],
            'Age':[23, 22, 23, 22, 18],
            'Days':[5, 8, 9, 6,3],
            'Company':['Apexon','QSG','Apexon','Apexon','info']
            }
    
    print("calling draw_line method")
    draw_line(data)

if __name__ == '__main__':
    main()
```
