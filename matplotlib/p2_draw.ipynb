```python
import matplotlib.pyplot as plt

def main():
 
    x = [10,30,50,70]
    y = [2,402,650,20]
    plt.plot(x,y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

if __name__ =='__main__':
    main()
```
