```python
import matplotlib.pyplot as plt

def plot_lines(x1, y1, x2, y2,x3, y3):
    """
    Plots two lines on the same plot with legends.
    
    Args:
        x1 (list): X-axis values for the first line.
        y1 (list): Y-axis values for the first line.
        x2 (list): X-axis values for the second line.
        y2 (list): Y-axis values for the second line.
    """
    plt.figure(figsize=(10, 6))

    # Plotting the first line
    plt.plot(x1, y1, marker='o', linestyle='-', color='b', label='Line 1')

    # Plotting the second line
    plt.plot(x2, y2, marker='s', linestyle='--', color='r', label='Line 2')
    
    # Plotting the third line
    plt.plot(x3, y3, marker='^', linestyle='-.', color='b', label='Line 3')
    

    # Adding labels and title
    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.title('Multiple Lines Plot')

    # Adding legend
    plt.legend()

    # Display the grid
    plt.grid(True)

    # Show the plot
    plt.show()

def main():
    # Sample data for the first line
    x1 = [1, 2, 3, 4, 5]
    y1 = [2, 4, 6, 8, 10]

    # Sample data for the second line
    x2 = [1, 2, 3, 4, 5]
    y2 = [1, 3, 5, 7, 9]
    
    # Sample data for the third line
    x3 = [9, 8, 7, 6, 5]
    y3 = [1, 3, 5, 7, 9]

    plot_lines(x1, y1, x2, y2,x3,y3)

if __name__ == '__main__':
    main()
```
