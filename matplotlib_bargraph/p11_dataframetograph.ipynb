```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Sample DataFrame
    data = {
        'a': [2, 4, 6, 8, 10],
        'b': [4, 2, 4, 2, 2],
        'c': [8, 3, 7, 6, 4],
        'd': [5, 4, 4, 4, 3],
        'e': [7, 2, 7, 8, 3],
        'f': [6, 6, 8, 6, 2]
    }
    indexing=(len(data['a']))
    df = pd.DataFrame(data)
    position=np.arange(indexing)
    
    # Set the index to column 'a' for easier plotting
    df.set_index(position, inplace=True)
    
    # Plot the data
    df.plot(kind='bar', figsize=(10, 6), width=0.8)

    # Add titles and labels
    plt.title('Bar Plot from DataFrame')
    plt.xlabel('Index (a)')
    plt.ylabel('Values')
    plt.legend(title='Columns')
    
    # Display the plot
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    # Sample data as a dictionary
    data = {
        'a': [2, 4, 6, 8, 10],
        'b': [4, 2, 4, 2, 2],
        'c': [8, 3, 7, 6, 4],
        'd': [5, 4, 4, 4, 3],
        'e': [7, 2, 7, 8, 3]
    }
    
    # Create DataFrame
    df = pd.DataFrame(data)
    
    # Plot the data
    df.plot(kind='bar', figsize=(10, 6), width=0.8)

    # Add titles and labels
    plt.title('Bar Plot from DataFrame')
    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.legend(title='Columns')
    
    # Display the plot
    plt.show()

if __name__ == "__main__":
    main()
```
