```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Sample data
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
    labels = ['Measure 1', 'Measure 2', 'Measure 3', 'Measure 4']

    # Convert mean_velocity to integer values for text labels
    mean_velocity_int = [int(round(value)) for value in mean_velocity]

    # Number of bars
    num_bars = len(mean_velocity)
    
    # X positions for the bars
    positions = np.arange(num_bars)
    
    # Bar width
    bar_width = 0.5
    
    # Create figure and axis
    plt.figure(figsize=(10, 6))
    
    # Plot bars with error bars
    bars = plt.bar(positions, mean_velocity, yerr=std_deviation, width=bar_width, color='skyblue', edgecolor='black', capsize=5, label='Mean Velocity')
    
    # Add titles and labels
    plt.title('Bar Plot with Error Bars and Text Labels')
    plt.xlabel('Measurements')
    plt.ylabel('Mean Velocity')
    plt.xticks(positions, labels)
    
    # Add text labels above each bar
    for bar, value in zip(bars, mean_velocity_int):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, str(value), ha='center', va='bottom', fontsize=10, color='black')
    
    # Add legend
    plt.legend()
    
    # Display the plot
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Sample data
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
    labels = ['Measure 1', 'Measure 2', 'Measure 3', 'Measure 4']

    # Convert mean_velocity to integer values for text labels
    mean_velocity_int = [int(round(value)) for value in mean_velocity]

    # Number of bars
    num_bars = len(mean_velocity)
    
    # X positions for the bars
    positions = np.arange(num_bars)
    
    # Bar width
    bar_width = 0.5
    
    # Create figure and axis
    plt.figure(figsize=(10, 6))
    
    # Plot bars with error bars
    bars = plt.bar(positions, mean_velocity, yerr=std_deviation, width=bar_width, color='skyblue', edgecolor='black', capsize=5, label='Mean Velocity')
    
    # Add titles and labels
    plt.title('Bar Plot with Error Bars and Text Labels')
    plt.xlabel('Measurements')
    plt.ylabel('Mean Velocity')
    plt.xticks(positions, labels)
    
    # Add text labels above each bar
    for bar, value in zip(bars, mean_velocity):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() , height + 0.02, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black')
    
    # Add legend
    plt.legend()
    
    # Display the plot
    plt.show()

if __name__ == "__main__":
    main()
```
