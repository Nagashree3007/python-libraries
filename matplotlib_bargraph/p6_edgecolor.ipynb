```python
import matplotlib.pyplot as plt

def main():
    '''
    
        Definition: Provides a brief description of what the function does and its purpose.
        
        Parameters: The function does not accept any arguments.
        
        Return Value: The function performs its tasks by creating and displaying the chart but does not return any value.
        
    '''
    # Data for the bar chart
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

    # Create the bar chart
    plt.figure(figsize=(10, 6))
    bars = plt.bar(programming_languages, popularity, color='skyblue', edgecolor='blue')

    # Add titles and labels
    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')

    # Add text labels above each bar
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval}', 
                 ha='center', va='bottom', fontsize=10, color='black')

    # Display the bar chart
    plt.show()

if __name__ == "__main__":
    main()
```
