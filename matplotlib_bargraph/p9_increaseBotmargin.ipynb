```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Data for the bar chart
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
    
    # Specify the positions and widths of each bar
    positions = np.arange(len(programming_languages))
    bar_width = 0.6

    # Create a new figure with a specific size
    plt.figure(figsize=(10, 6))

    # Create the bar chart
    bars = plt.bar(positions, popularity, width=bar_width, color='skyblue', edgecolor='blue')

    # Add titles and labels
    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')

    # Set the x-ticks to be at the positions of the bars
    plt.xticks(positions + bar_width / 2, programming_languages)

    # Increase bottom margin
    plt.subplots_adjust(bottom=0.2)  # Adjust the bottom margin (0.2 is a fraction of the figure height)

    # Add text labels above each bar
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval}', 
                 ha='center', va='bottom', fontsize=10, color='black')

    # Display the bar chart
    plt.show()

if __name__ == "__main__":
    main()
```
