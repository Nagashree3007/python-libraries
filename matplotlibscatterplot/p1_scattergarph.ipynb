```python
import matplotlib.pyplot as plt
import numpy as np

# Set the seed for reproducibility
np.random.seed(0)

# Generate random data
num_points = 100  # Number of points to plot
x = np.random.rand(num_points) * 100  # Random values for X axis
print(x)

y = np.random.rand(num_points) * 100  # Random values for Y axis
print(y)

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='blue', marker='o', alpha=0.7)

# Add labels and title
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot of Random Data')

# Show the plot
plt.grid(True)
plt.show()
```

#using randint keyword

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the seed for reproducibility
np.random.seed(0)

# Generate random data
num_points = 100  # Number of points to plot
x = np.random.randint(num_points,size=10)   # Random values for X axis
print(x)

y = np.random.randint(num_points,size=10)   # Random values for Y axis
print(y)

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='blue', marker='o', alpha=0.7)

# Add labels and title
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot of Random Data')

# Show the plot
plt.grid(True)
plt.show()
```
