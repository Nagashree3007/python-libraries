```python
import matplotlib.pyplot as plt
import numpy as np
import random
groups=['Group1','group2','group3']
x=[np.random.rand(1,5) for _ in range(3)]
y=[np.random.rand(1,5) for _ in range(3)]
z={}
for i in range (3):
    z[i]=[x[i],y[i]]  
    

c=["#1f77b4", "#ff7f0e", "#2ca02c"]   
s=['s','o','^'] 

for key ,val in z.items():
    label='Group'+str(key)
    plt.scatter(val[0], val[1], marker=s[key], label=label , color=c[key])
    
# Add labels and title
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('Scatter Plot of Weights and Heights for Three Different Groups')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the seed for reproducibility
np.random.seed(0)

# Generate sample data
num_points = 50  # Number of points per group

# Group 1
weights1 = np.random.normal(60, 10, num_points)  # Mean weight: 60 kg, Std dev: 10 kg
heights1 = np.random.normal(165, 7, num_points)   # Mean height: 165 cm, Std dev: 7 cm

# Group 2
weights2 = np.random.normal(70, 12, num_points)  # Mean weight: 70 kg, Std dev: 12 kg
heights2 = np.random.normal(175, 8, num_points)   # Mean height: 175 cm, Std dev: 8 cm

# Group 3
weights3 = np.random.normal(80, 15, num_points)  # Mean weight: 80 kg, Std dev: 15 kg
heights3 = np.random.normal(180, 10, num_points)  # Mean height: 180 cm, Std dev: 10 cm

# Create the scatter plot
plt.figure(figsize=(10, 6))

# Plot each group with different colors and markers
plt.scatter(weights1, heights1, color='red', marker='o', label='Group 1')
plt.scatter(weights2, heights2, color='blue', marker='s', label='Group 2')
plt.scatter(weights3, heights3, color='green', marker='^', label='Group 3')

# Add labels and title
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('Scatter Plot of Weights and Heights for Three Different Groups')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
