<h1>Q1.</h1>Write a program to draw a scatter plot for random 1000 x and y coordinates 

info on random functions:<br>

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
x = np.random.randn(1000)
y = np.random.randn(1000)

# Create scatter plot
fig = go.Figure(data=go.Scatter(
    x=x,
    y=y,
    mode='markers',  # Use 'markers' for a scatter plot
    marker=dict(size=7, color='red')  # Marker properties
))

# Update layout for better readability
fig.update_layout(
    title='Scatter Plot of Random Data',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show plot
fig.show()
```

<h1>Q2.</h1> Write a program to draw line and scatter plots for random 100 x and y coordinates 

```python
import plotly.graph_objects as go
import numpy as np
x=np.random.permutation(100)
y=np.random.permutation(100)

fig = go.Figure(go.Scatter(
    x=x,
    y=y,
    mode='markers',
    marker=dict(color='red',size=10),
    text=[f'point{i} {x[i]},{y[i]}' for i in range(100)],
    hoverinfo='text',
))

fig.update_layout(
    title='Scatter Plot of Random Data',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show plot
fig.show()
```

<h1>Q3.</h1> Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.random.rand(500)
y = np.random.rand(500)

# Create the scatter plot
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode='markers',  # Scatter plot
    marker=dict(
        size=10,  # Size of the markers
        color=np.random.rand(500),  # Color based on random values
        colorscale='Viridis',  # Color scale
        opacity=0.8  # Marker opacity
    ),
    text=[f'Point {i}' for i in range(500)],  # Hover text
    hoverinfo='text',  # Display hover text
))       

# Update layout with titles and axis labels
fig.update_layout(
    title='Scatter Plot of 500 Random Points',
    xaxis_title='X Axis',
    yaxis_title='Y Axis',
    plot_bgcolor='black',  # Background color of the plot area
    paper_bgcolor='white',  # Background color of the paper
    margin=dict(l=40, r=20, t=40, b=20)  # Margins
)

# Show the plot
fig.show()
```

<h1>Q4.</h1> Write a program to draw a scatter plot for a given dataset and show datalabels on hover 
https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

```python
import plotly.express as px
import pandas as pd

# Load the dataset from the provided URL
url = 'https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv'
df = pd.read_csv(url)

# Create the scatter plot
fig = px.scatter(
    df,
    x='Rank',
    y='Population',
    text='State',  # Data labels to display on hover
    title='US States by Population (2014)',
    labels={'Rank': 'Rank', 'Population': 'Population'},
    color='Population',  # Color by population size 
    color_continuous_scale='Viridis'  # Color scale
)

# Update the hover information
fig.update_traces(
    textposition='top center',  # Position of the text labels
    marker=dict(size=15, opacity=0.8),  # Marker size and opacity
    selector=dict(mode='markers+text')  # Show both markers and text
)

# Update layout for better presentation
fig.update_layout(
    plot_bgcolor='rgba(240, 240, 240, 0.8)',  # Background color of the plot area
    paper_bgcolor='white',  # Background color of the paper
    margin=dict(l=40, r=20, t=40, b=20),  # Margins
    xaxis_title='Rank',
    yaxis_title='Population'
)

# Show the plot
fig.show()
```
