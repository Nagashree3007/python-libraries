<h1>Q1.</h1> Write a program to draw a point plot for sex against survived for a dataset given in url 
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
data = pd.read_csv(url)

# Create the point plot
plt.figure(figsize=(8, 6))
sns.pointplot(x='sex', y='survived', data=data,errorbar=None, markers='o')

# Customize the plot
plt.title('Point Plot of Sex Against Survival')
plt.xlabel('Sex')
plt.ylabel('Survived')
plt.ylim(-0.1, 1.1)  # To better visualize the binary outcome

# Show the plot
plt.show()
```
