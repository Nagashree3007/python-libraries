<h1>Q1.</h1>Write a program to draw a swarm plot of total bill against size  for a  given dataset 
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
data = pd.read_csv(url)

# Create the swarm plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='size', y='total_bill', data=data,hue='size',palette='viridis', size=7)

# Customize the plot
plt.title('Swarm Plot of Total Bill Against Size')
plt.xlabel('Size')
plt.ylabel('Total Bill')

# Show the plot
plt.show()
```

<h1>Q2.</h1> Write a program to draw swarm plot of “total bill” against day for a dataset given in url 
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
data = pd.read_csv(url)

# Create the swarm plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='day', y='total_bill', data=data,hue='day',palette='viridis', size=7)

# Customize the plot
plt.title('Swarm Plot of Total Bill Against Size')
plt.xlabel('day')
plt.ylabel('Total Bill')

# Show the plot
plt.show()
```
